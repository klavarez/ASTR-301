```python
import numpy as np 
import matplotlib.pyplot as plt
import astropy.units as u
from astropy.coordinates import SkyCoord
```

```python
# importing the data
hyades = './HyadesDR3.txt'
hyadesData = np.loadtxt(hyades, skiprows=2, dtype=float)
data = []

# separating each row of data
for rows in hyadesData:
    data.append(rows)

# data type arrays
raICRS, deICRS = [], []     # right ascension and declination [deg]
plx, ePlx = [], []          # parallax and its error [mas]
pmRA, epmRA = [], []        # right ascension proper motion & error [mas/yr]
pmDE, epmDE = [], []        # declination proper motion & error [mas/yr]
Gmag, eGmag = [], []        # gband magnitude & error [mag]
BPmag, eBPmag = [], []      # bpmagnitude & error [mag]
RPmag, eRPmag = [], []      # rpmagnitude & error [mag]

# separating each value in every row by datatype
for i in range(len(data)):
    raICRS.append(float(data[i][0]))
    deICRS.append(float(data[i][1]))
    plx.append(float(data[i][2]))
    ePlx.append(float(data[i][3]))
    pmRA.append(float(data[i][4]))
    epmRA.append(float(data[i][5]))
    pmDE.append(float(data[i][6]))
    epmDE.append(float(data[i][7]))
    Gmag.append(float(data[i][8]))
    eGmag.append(float(data[i][9]))
    BPmag.append(float(data[i][10]))
    eBPmag.append(float(data[i][11]))
    RPmag.append(float(data[i][12]))
    eRPmag.append(float(data[i][13]))
```

```python
# FILTERING THE DATA - PARALLAX HIST #
# plotting a histogram of this cluster by plotting the parallax data
bins = np.linspace(-20, 40, 100)
plt.hist(plx, bins, log=True, label='All')
plt.xlabel('Parallax [mas]')
plt.ylabel('Number of objects')
plt.title('Hyades GaiaDR3 Raw Data')
plt.legend()
plt.show()
# theres a bump from 15.4 to 25
```

```python
# FILTERING THE DATA - PROPER MOTION SCATTER #
plt.scatter(pmRA, pmDE, s=0.5)
# zoom in on densest section of data
plt.axis([250, -125, -300, 200])
plt.title('Hyades GaiaDR3 Raw Data')
plt.xlabel('PM RA [mas/yr]')
plt.ylabel('PM DE [mas/yr]')
plt.show()
# cluster at pmra=78.8 pmde=-44.9 to pmra=124.0 pmde=-12.0
```

```python
# PLOTTING ISOCHRONES #
def readIsochrone(filename): # reading in an isochrone
    # open text file
    file = np.loadtxt(filename, skiprows=12)
    data = []

    # separating each row of data
    for rows in file:
        data.append(rows)
    
    # data type arrays
    logAge = []     # age of the isochrone [yr]
    Gmag = []       # g-band magnitude of isochrone [mag]
    GRPmag = []     # g-rp band magnitude (color) of the isochrone [mag]
    GBPmag = []     # g-bp band magnitude (color) of the isochrone [mag]
    
    # only taking important stuff (age, gmag, bpmag, rpmag)
    for i in range(len(data)):
        logAge.append(float(data[i][2]))
        Gmag.append(float(data[i][28]))
        GRPmag.append(float(data[i][30]))
        GBPmag.append(float(data[i][29]))
    
    # isolating the RPmag and BPmag
    RPmag = [GRPmag[i] - Gmag[i] for i in range(len(GRPmag))]
    BPmag = [GBPmag[i] - Gmag[i] for i in range(len(GBPmag))]
    # subtracting them from each other for the color
    BRmag = [BPmag[i] - RPmag[i] for i in range(len(BPmag))]

    # # undoing the log of the age
    # estAge = []
    # for i in range(len(logAge)):
    #     val = 10**logAge[i]
    #     estAge.append(float(val))
    
    return logAge, Gmag, BRmag
```

```python
# function to plot an isochrone
def isoPlot(filename):
    age, Gmag, BRmag = readIsochrone(filename)
    # plotting the diagram
    plt.plot(BRmag, Gmag)
    plt.xlabel('B - R [mag]')
    plt.ylabel('G [mag]')
    # plt.gca().invert_xaxis()
    # plt.gca().invert_yaxis()
    
```

```python
# EDITING AN ISOCHRONE TO FIT THE DATA #
# plt.scatter(clustercolorPMPX, Gmag3, s=1, label='Filtered data') # first plotting the isochrone
# plt.gca().invert_xaxis()

def editIsochrone(filename, shiftx, shifty, flipx, flipy, color):   # fitting an isochrone & editing it to fit the data
    # run readIsochrone function
    age, Gmag, BRmag = readIsochrone(filename)
    
    # if flipy is negative, flip magnitude over the x axis
    # shift magnitude up or down by shifty
    shiftedG = [flipy * (i + shifty) for i in Gmag]
    # if flipx is negative, flip color over the y axis
    # shift color left or right by shiftx
    shiftedColor = [flipx * (i + shiftx) for i in BRmag]
    
    # plot ischrone with shifted values
    plt.plot(shiftedColor, shiftedG, linewidth=2, alpha=0.5, label=f'log age: {age[0]}', color=color)
    plt.xlabel("B - R [mag]")
    plt.ylabel("G [mag]")
    # plt.title(filename.replace(".txt", ""))
    plt.legend()

# # plotting an isochrone against the filtered data
# editIsochrone('hyadesIso7.txt', 1.5, 12, 1, 1, 'darkviolet')
# # plt.axis([4.5, -1.5, 0, 30])
# plt.gca().invert_yaxis()
# plt.gca().invert_xaxis()
```
